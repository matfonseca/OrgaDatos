```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

%matplotlib inline
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
from sklearn.base import TransformerMixin

class DataFrameImputer(TransformerMixin):

    def __init__(self):
        """Impute missing values.

        Columns of dtype object are imputed with the most frequent value 
        in column.

        Columns of other types are imputed with mean of column.

        """
    def fit(self, X, y=None):

        self.fill = pd.Series([X[c].value_counts().index[0]
            if X[c].dtype == np.dtype('O') else X[c].mean() for c in X],
            index=X.columns)

        return self

    def transform(self, X, y=None):
        return X.fillna(self.fill)
```

```python
#Mapeo los valores de nuestro analisis de datos para el Random Forest
postulations = pd.read_csv('../DataAnalysis/postulaciones_nopostulaciones.csv')
```

```python
postulations = DataFrameImputer().fit_transform(postulations)
```

```python
postulations.head()
```

```python
postulations = postulations.iloc[:,3:]
```

```python
#Transformamos los datos categóricos en números
for column in postulations.columns:
    if postulations[column].dtype == type(object):
        postulations[column] = le.fit_transform(postulations[column].astype(str))
```

```python
#Normalizamos los valores
scaler = MinMaxScaler() 
scaled_values = scaler.fit_transform(postulations) 
postulations.loc[:,:] = scaled_values
```

```python
postulations.head()
```

```python
postulations.to_csv('DataMappedRandomForest2.csv', index=False)
```
