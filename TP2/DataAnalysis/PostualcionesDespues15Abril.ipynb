```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def datetime(date,format_string):
    return pd.to_datetime(date,format= format_string,errors='coerce')
```

```python
data_postulation1= pd.read_csv('../Data/datos_navent_fiuba/fiuba_4_postulaciones.csv')
data_postulation1.head()
```

```python
data_postulation2= pd.read_csv('../Data/Hasta15Abril/fiuba_4_postulaciones.csv')
data_postulation2.head()
```

```python
#Concateno los dos periodos de postulaciones
frames_postulation = [data_postulation1,data_postulation2]
data_postulation = pd.concat(frames_postulation, ignore_index = True)
data_postulation.drop_duplicates(inplace = True, keep = 'first')
data_postulation.reset_index(drop=True, inplace = True)
```

```python
data_postulation['fechapostulacion'] = datetime(data_postulation['fechapostulacion'],'%Y-%m-%d')
```

```python
data_postulation['Mayor15Abril'] = data_postulation['fechapostulacion'].map(lambda x: (x.month >= 4) and(x.day>15))
```

```python
data_postulation['Mayor15Abril'].value_counts()
```

```python
data_postulation = data_postulation[data_postulation['Mayor15Abril'] == True]
```
