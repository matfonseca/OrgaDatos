```python
total.to_csv('pruebaDescripciones.csv')
```

```python
import re

def cleanhtml(raw_html):
    cleanr = re.compile('<.*?>')
    subclean = re.sub(cleanr, '', raw_html)
    cleantext = subclean.replace("\r","")
    cleantext = cleantext.replace('\n',"")
    return cleantext
```

```python
total['descripcion'] = total["descripcion"].map(lambda x: cleanhtml(x))
```
